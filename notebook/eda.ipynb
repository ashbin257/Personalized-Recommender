```python
import pandas as pd

# Load the dataset
df = pd.read_csv("C:/Users/ashbi/OneDrive/Desktop/tech/personalized-recommender/data/Reviews.csv")

# Preview the dataset
print("Shape:", df.shape)
df = df[['UserId', 'ProductId', 'Score', 'Summary', 'Text']].dropna()
print(df.head())
```

```python
print("Shape:", df.shape)
print(df.describe())
print("Shape:", df.shape)
print("Columns:", df.columns)
```

```python
missing = df.isnull().sum()
print("Missing values:\n", missing)
```

```python
# Unique users and products
print("Unique Users:", df['UserId'].nunique())
print("Unique Products:", df['ProductId'].nunique())
```

```python
top_users = df['UserId'].value_counts().head(10)
print("Top reviewers:\n", top_users)
```

```python
top_products = df['ProductId'].value_counts().head(10)
print("Most reviewed products:\n", top_products)
```

```python
# Rating distribution
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(data=df, x='Score')
plt.title("Rating Distribution")
plt.xlabel("Score")
plt.ylabel("Count")
plt.show()
```

```python
df['review_length'] = df['Text'].apply(lambda x: len(str(x).split()))
sns.histplot(df['review_length'], bins=50)
plt.title("Distribution of Review Lengths")
plt.xlabel("Number of Words in Review")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(x='Score', y='review_length', data=df)
plt.title("Review Length vs Rating")
plt.xlabel("Rating")
plt.ylabel("Review Length")
plt.show()
```
